```python
import time

def time_counter_for_dataset_last(n_samples = 1000):
    # Get the start time
    start_time = time.perf_counter()
    n_samples_in_dataset = len(LunaDataset())
    for i in range(n_samples_in_dataset - n_samples, n_samples_in_dataset-1):
        dummy_inst = LunaDataset()[i]
    # Get the end time
    end_time = time.perf_counter()
    # Calculate the execution time
    execution_time = end_time - start_time
    print("Execution time:", execution_time, "seconds")
    return execution_time

```

```python
import time

def time_counter_for_dataset(n_samples = 1000):
    # Get the start time
    start_time = time.perf_counter()
    for i in range(n_samples):
        dummy_inst = LunaDataset()[i]
    # Get the end time
    end_time = time.perf_counter()
    # Calculate the execution time
    execution_time = end_time - start_time
    print("Execution time:", execution_time, "seconds")
    return execution_time
```
